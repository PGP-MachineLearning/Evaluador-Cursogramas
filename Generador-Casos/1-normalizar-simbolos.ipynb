Parametros


```
max_size = (100, 100)
path_origen = 'gdrive/My Drive/Investigaciones/Cursogramas/simbolos originales/'
path_destino = 'gdrive/My Drive/Investigaciones/Cursogramas/simbolos/'
```

Librerias

```
import os
import os.path
from os.path import isfile, join
from PIL import Image, ImageChops
```

Normalizar imagenes

```
# crea el directorio destino, si es necesario
if not os.path.isdir(path_destino):
  os.makedirs(path_destino)

simbolos = [file for file in os.listdir(path_origen) if isfile(join(path_origen, file))]

for simbolo in simbolos:
  imagen = Image.open(join(path_origen, simbolo))

  bg = Image.new(imagen.mode, imagen.size, imagen.getpixel((0, 0)))
  diff = ImageChops.difference(imagen, bg)
  diff = ImageChops.add(diff, diff, 2.0, -100)
  box = diff.getbbox()
  if box:
    imagen = imagen.crop(box)

  imagen.thumbnail(max_size, Image.ANTIALIAS)
  imagen.save(join(path_destino, simbolo), 'PNG')
```
